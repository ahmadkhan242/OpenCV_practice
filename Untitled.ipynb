```python
from __future__ import print_function 
import argparse 
import cv2 
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

# Loading Image

```python
img = cv2.imread("./OpenCV_practice/pic.jpg")
```

```python
plt.imshow(img)
```

# Creating Line on image  

```python
canvas = np.zeros((300, 300, 3), dtype="uint8")
```

```python
green = (0, 255, 0)
cv2.line(canvas, (0,0), (300, 300), green)
cv2.line(canvas, (0,300), (300, 0), (255, 0 ,0))
```

```python
plt.imshow(canvas)
```

# Creating Rectangle

```python
# args(image, starting point, ending point, color, thickness)
cv2.rectangle(canvas, (10, 10), (60, 60), green, 5)
plt.imshow( canvas)
```

# Creating Circle 

```python
circle_canvas = np.zeros((300, 300, 3), dtype="uint8")
(cX, cY) = (circle_canvas.shape[1]//2, circle_canvas.shape[0]//2)
white = (255, 255, 255)
for r in range(0, 175, 25):
    cv2.circle(circle_canvas, (cX, cY), r, white)
    
plt.imshow(circle_canvas)
```

```python
drawing_canvas = np.zeros((300, 300, 3), dtype="uint8")
for i in range(0, 55):
    radius = np.random.randint(5, high = 200)
    color = np.random.randint(0, high = 256, size = (3,)).tolist ()
    pt = np.random.randint(0, high = 300, size = (2,))
    cv2.circle(drawing_canvas, tuple(pt), radius, color, -1)
plt.imshow(drawing_canvas)
```

# Shifting Image

```python
# Library that hendels alot of things in openCv
import imutils
M = np.float32([[1, 0, 25], [0, 1, 50]])
shifted = cv2.warpAffine(img, M, (img.shape[1], img.shape [0]))
shifted = imutils.translate(img, 0, 100)
plt.imshow(shifted)
# cv2.imshow("Shifted Down", shifted)
# cv2.waitKey(0)
```

# Rotating Image

```python
image = cv2.imread("./OpenCV_practice/pic.jpg")
(h, w) = image.shape[:2] 
center = (w // 2, h // 2) 
M = cv2.getRotationMatrix2D(center, 45, 1.0) 
rotated = cv2.warpAffine(image, M, (w, h)) 
rotated = imutils.rotate(image, 90)
plt.imshow(rotated)
```

# Resizing Image

```python
r = 50.0 / image.shape[0] 
dim = (int(image.shape[1] * r), 50)
resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)#Option for interpolation(INTER_LINEAR, INTER_CUBIC, INTER_NEAREST)
plt.imshow(resized)
cv2.imshow("Resize", resized)
cv2.waitKey(0)
```

# Filliping A Image

```python
image = cv2.imread("./OpenCV_practice/pic.jpg")
plt.imshow(image)
plt.title('Original')
```

```python
image = cv2.imread("./OpenCV_practice/pic.jpg")
Flipped_Horizontally = cv2.flip(image, 1)
plt.imshow(Flipped_Horizontally)
plt.title('Flipped_Horizontally')
```

```python
image = cv2.imread("./OpenCV_practice/pic.jpg")
Flipped_Vertically = cv2.flip(image, 0)
plt.imshow(Flipped_Vertically)
plt.title('Flipped_Vertically')
```

```python
image = cv2.imread("./OpenCV_practice/pic.jpg")
Flipped_Vertically_and_Horizontally = cv2.flip(image, -1)
plt.imshow(Flipped_Vertically_and_Horizontally)
plt.title('Flipped_Vertically_and_Horizontally')
```

# Croping Image

```python
image = cv2.imread("./OpenCV_practice/pic.jpg")
cropped = image[30:120 , 240:335]
plt.imshow(cropped)
plt.title('cropped')
```

# Arithmetic opration on image

```python
image = cv2.imread("./OpenCV_practice/pic.jpg")
M = np.ones(image.shape, dtype = "uint8") * 100 
added = cv2.add(image, M)
plt.imshow(added)
```

```python
image = cv2.imread("./OpenCV_practice/pic.jpg")
M = np.ones(image.shape, dtype = "uint8") * 100 
subtracted = cv2.subtract(image, M)
plt.imshow(subtracted)
```
